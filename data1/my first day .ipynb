```python
import pandas as pd
```

```python
user_cols = ['user_id','age', 'gender', 'occupation', 'zip_code']
users = pd.read_table('u.user', sep='|', header = None,names = user_cols, index_col = 'user_id')
users.head()
```

## removing duplicates

```python
# check duplicates when there is one it will be true
users.zip_code.duplicated().head()
```

```python
#counting duplicates items (treu becomes 1 and false becomes 0)
users.zip_code.duplicated().sum() # deuplicated zipcodes
```

```python
#detect duplicates dataframe (rows: if an entire row is duplicated identical to a previous row )
users.duplicated().head()
```

```python
users.duplicated().sum()
```

```python
#keep = first ( means ignore the first  )
```

```python
#examine the duplicate rows by ignoring the first occurrence
users.loc[users.duplicated(keep='first'),:]
```

```python
#examine the duplicate rows by ignoring the last occurrence
users.loc[users.duplicated(keep='last'),:]
```

```python
#examine the duplicate rows by by marking all occurrences

users.loc[users.duplicated(keep=False),:]
```

```python
users2 = users.copy()
users.drop_duplicates(keep='last').shape
```

```python
users.shape
```

```python
# if you want to drop them permanently
users.drop_duplicates(keep='last', inplace = True)
```

#applying functions to a data frame

```python
train = pd.read_csv('titanic_train.csv')
train.head()
```

```python
#map 'female' to 0 and 'male' to 1
train['Sex_num'] = train.Sex.map({'female':1,'male':0})
```

```python
train[['Sex_num','Sex']].head(4)
```

```python
train.loc[0:4,['Sex_num','Sex']]
```

```python
train['Name_length']= train.Name.apply(len)
```

```python
train.head()
```

```python
train.loc[0:4, ['Name','Name_lenght']]
```

```python
#round up each element in the 'fare' series to the next in 
import numpy as np
train['Fare_ceil']= train.Fare.apply(np.ceil)
train.loc[0:4,['Fare','Fare_ceil']]
```

```python
# floor rounds down
train['Fare_ceil']= train.Fare.apply(np.floor)
train.loc[0:4,['Fare','Fare_ceil']]
```

```python
# way of counting females in dataframe
a = train.Sex
count=0
for i in  a:
    if i== 'female':
        count+=1
print(count)
    
    
```

```python
# way of counting females in dataframe by using value_counts
train.Sex.value_counts()
```
