# grids

```python
iris = pd.read_csv('iris.csv')
iris.head()
```

```python
iris.species.unique()
```

# pair grid

```python
# JUST Gids
sns.PairGrid(iris)
```

```python
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
# map to upper , lower, and diagono
g = sns.PairGrid(iris)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)

```

```python
sns.pairplot(iris,hue = 'species',palette='rainbow')
```

### box plots and violin plots
###### box plots and violin plots are used to show the distribution of categorical data


```python
# distribution of days in relation to the total bill column
sns.boxplot(x='day',y='total_bill',data=tips,palette='rainbow')
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,palette='coolwarm')
```

```python
sns.boxplot(x='day',y='total_bill',hue = 'smoker',data=tips,palette='coolwarm')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,palette='coolwarm')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue = 'sex',palette='Set1',scale='count')
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter= False)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter= True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter= True,hue='sex',palette='Set1')
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter= True,hue='time',palette='Set1')
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter= True,hue='smoker',palette='Set1')
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter= True,hue='smoker',palette='Set1',split = True)
```

### swarmplot

```python
sns.swarmplot(x='day',y='total_bill',data=tips,hue='smoker',palette='Set1',split= True)
```
